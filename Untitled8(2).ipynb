# **Regression simple**

```python
import pandas as pd
import seaborn as sns
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
```

```python
data=pd.read_csv('Student-Attitude-and-Behavior.csv',sep=",")
data.info()
```

```python
data.head()
```

```python
data['Certification Course'] = data['Certification Course'].replace({'Yes': 1, 'No': 0})
data['part-time job'] = data['part-time job'].replace({'Yes': 1, 'No': 0})
# Séparation des données en X et Y
X = data[['part-time job','daily studing time']]
Y = data[['12th Mark']]

# Séparation des données en ensembles d'entraînement et de test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Construction du modèle de régression linéaire
model = LinearRegression()
model.fit(X_train, Y_train)

train_score = model.score(X_train, Y_train)
test_score = model.score(X_test, Y_test)

print(f"Score d'entraînement : {train_score}")
print(f"Score de test : {test_score}")
```

```python
Y_pred = model.predict(X_test)

# Création du graphique
plt.figure(figsize=(6, 6))
plt.scatter(Y_test, Y_pred, color='blue', label='Prédictions')
plt.plot(Y_test, Y_test, color='red', linewidth=2, label='Valeurs réelles')
plt.xlabel('Valeurs réelles (12th Mark)')
plt.ylabel('Prédictions')
plt.title('Régression linéaire - Prédictions vs Valeurs réelles')
plt.legend()
plt.show()
```

# **Interpretation**

D'après les résultats, on peut dire qu'il existe une relation positive entre le nombre d'heures d'étude quotidiennes d'un étudiant et sa note du bac, c'est-à-dire que plus le nombre d'heures est grand, plus la note est élevée.

Donc, il y a une corrélation positive entre les deux variables X et Y.

# **Anova**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
sns.set()

ax = sns.boxplot(x="part-time job", y="12th Mark", data=data, color='white')
plt.xlabel('part-time job')
plt.ylabel('marks')
plt.title('Boîtes à moustaches')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.set()

ax = sns.boxplot(x="prefer to study in", y="12th Mark", data=data, color='white')
plt.xlabel('prefer to study in')
plt.ylabel('marks')
plt.title('Boîtes à moustaches')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.set()

ax = sns.boxplot(x="Financial Status", y="12th Mark", data=data, color='white')
plt.xlabel('Financial Status')
plt.ylabel('marks')
plt.title('Boîtes à moustaches')
plt.show()
```

```python
# Renommer les colonnes
data = data.rename(columns={'12th Mark': 'Twelfth_Mark', 'Financial Status': 'Financial_Status'})

# Afficher les noms de colonnes pour vérification
print(data.columns)

# Effectuer l'ANOVA avec la bonne formule
anova_variete = smf.ols('Twelfth_Mark ~ Financial_Status', data=data).fit()

# Afficher les résultats de l'ANOVA
sm.stats.anova_lm(anova_variete)
```

```python
plt.figure(figsize=(10, 6))
sns.set()

ax = sns.boxplot(x="social medai & video", y="12th Mark", data=data, color='white')
plt.xlabel('social medai & video')
plt.ylabel('marks')
plt.title('Boîtes à moustaches')
plt.show()
```

```python
# Rename columns
data = data.rename(columns={'12th Mark': 'Twelfth_Mark', 'social medai & video': 'Social_Media_Video'})

# Perform ANOVA
anova_variete = smf.ols('Twelfth_Mark ~ Social_Media_Video', data=data).fit()

sm.stats.anova_lm(anova_variete)
```

# **ANOVA a deux facteurs**

```python
data = data.rename(columns={'part-time job': 'part_time_job'})
anova_variete_phyto = smf.ols('Twelfth_Mark~part_time_job+Financial_Status', data=data).fit()
sm.stats.anova_lm(anova_variete_phyto)
```

```python
anova_variete_phyto = smf.ols('Twelfth_Mark~part_time_job*Financial_Status', data=data).fit()
sm.stats.anova_lm(anova_variete_phyto)
```

```python
data = data.rename(columns={'prefer to study in': 'prefer_to_study_in'})
anova_variete_phyto = smf.ols('Twelfth_Mark~prefer_to_study_in+Social_Media_Video', data=data).fit()
sm.stats.anova_lm(anova_variete_phyto)
```

```python
anova_variete_phyto = smf.ols('Twelfth_Mark~prefer_to_study_in*Social_Media_Video', data=data).fit()
sm.stats.anova_lm(anova_variete_phyto)
```

```python
import pandas as pd
from fanalysis.ca import CA
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

```python
data_subset = data[['Financial_Status', 'Stress Level ']]

# Encodage one-hot pour les variables catégorielles
encoder = OneHotEncoder()
data_encoded = encoder.fit_transform(data_subset)

# Créer une instance de l'Analyse Factorielle des Correspondances (AFC)
afc_subset = CA(row_labels=data_subset.index.values, col_labels=encoder.get_feature_names_out(), stats=True)

# Effectuer l'AFC sur le sous-ensemble de données encodé
afc_subset.fit(data_encoded.toarray())

# Définir la taille de la figure directement avant de tracer le graphique
plt.rcParams["figure.figsize"] = (10, 6)  # Taille personnalisée de la figure

# Créer le graphique de l'AFC
afc_subset.mapping_row(num_x_axis=1, num_y_axis=2)
plt.show()
```
