# **Regression simple**

```python
import pandas as pd
import seaborn as sns
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
```

```python
data=pd.read_csv('Student-Attitude-and-Behavior.csv',sep=",")
data.info()
```

```python
data.head()
```

```python
data['Certification Course'] = data['Certification Course'].replace({'Yes': 1, 'No': 0})
data['part-time job'] = data['part-time job'].replace({'Yes': 1, 'No': 0})
# Séparation des données en X et Y
X = data[['part-time job','daily studing time']]
Y = data[['12th Mark']]

# Séparation des données en ensembles d'entraînement et de test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Construction du modèle de régression linéaire
model = LinearRegression()
model.fit(X_train, Y_train)

train_score = model.score(X_train, Y_train)
test_score = model.score(X_test, Y_test)

print(f"Score d'entraînement : {train_score}")
print(f"Score de test : {test_score}")
```

```python
Y_pred = model.predict(X_test)

# Création du graphique
plt.figure(figsize=(6, 6))
plt.scatter(Y_test, Y_pred, color='blue', label='Prédictions')
plt.plot(Y_test, Y_test, color='red', linewidth=2, label='Valeurs réelles')
plt.xlabel('Valeurs réelles (12th Mark)')
plt.ylabel('Prédictions')
plt.title('Régression linéaire - Prédictions vs Valeurs réelles')
plt.legend()
plt.show()
```

# **Interpretation**

D'après les résultats, on peut dire qu'il existe une relation positive entre le nombre d'heures d'étude quotidiennes d'un étudiant et sa note du bac, c'est-à-dire que plus le nombre d'heures est grand, plus la note est élevée.

Donc, il y a une corrélation positive entre les deux variables X et Y.

```python
import pandas as pd
from fanalysis.ca import CA
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

```python
data_subset = data[['Financial_Status', 'Stress Level ']]

# Encodage one-hot pour les variables catégorielles
encoder = OneHotEncoder()
data_encoded = encoder.fit_transform(data_subset)

# Créer une instance de l'Analyse Factorielle des Correspondances (AFC)
afc_subset = CA(row_labels=data_subset.index.values, col_labels=encoder.get_feature_names_out(), stats=True)

# Effectuer l'AFC sur le sous-ensemble de données encodé
afc_subset.fit(data_encoded.toarray())

# Définir la taille de la figure directement avant de tracer le graphique
plt.rcParams["figure.figsize"] = (10, 6)  # Taille personnalisée de la figure

# Créer le graphique de l'AFC
afc_subset.mapping_row(num_x_axis=1, num_y_axis=2)
plt.show()
```

```python
import pandas as pd
import prince

# Importation de données
df = pd.read_csv("Student-Attitude-and-Behavior.csv")

# prétraitement de base de donnée
indices_a_supprimer = [3,4,5,6,7,9,11,13,14,15]
df_categorical = df.drop(df.columns[indices_a_supprimer], axis=1)

#df_categorical = df.drop(columns=['Height(CM)', 'Weight(KG)','10th Mark','12th Mark','college mark','daily studing time','salary expectation'])
df_categorical
```

```python

import matplotlib.pyplot as plt
# Initialisation de l'ACM
mca = prince.MCA(n_components=2, random_state=42)

# Ajustement de l'ACM sur les données

mca = mca.fit(df_categorical)

# Transformation des données
transformed_data = mca.transform(df_categorical)

# Transformation des données
transformed_data = mca.transform(df_categorical)

# Affichage des résultats
print("\nCoordonnées des individus sur les premiers axes factoriels :\n", transformed_data)

# Contributions des variables
variable_contributions = mca.column_coordinates(df_categorical)
print("\nCoordonnées des variables sur les premiers axes factoriels :\n", variable_contributions)
#print("\nContributions des variables aux axes factoriels :\n", mca.column_correlations(df_categorical))

# Visualisation des résultats
#mca.plot_rows(show_points=True)
# Visualisation des résultats avec matplotlib
fig, ax = plt.subplots()

# Plot des individus
ax.scatter(transformed_data[0], transformed_data[1], alpha=0.5, label='Individus')
for i, txt in enumerate(df.index):
    ax.annotate(txt, (transformed_data.iloc[i, 0], transformed_data.iloc[i, 1]))

# Plot des variables
var_coords = mca.column_coordinates(df_categorical)
ax.scatter(var_coords[0], var_coords[1], alpha=0.5, color='black', label='Variables')
for i, txt in enumerate(var_coords.index):
    ax.annotate(txt, (var_coords.iloc[i, 0], var_coords.iloc[i, 1]), color='black')

ax.legend()
plt.title('ACM: Indices et Variables')
plt.xlabel('Composante 1')
plt.ylabel('Composante 2')
plt.grid(True)
plt.show()
```

# **Clustering K-means**

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
print(kmeans.cluster_centers_)
```
